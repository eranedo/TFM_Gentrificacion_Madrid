```python
import pandas as pd
import numpy as np
import os

RAW = '../data/raw/'
PROCESSED = '../data/processed/'
FINAL = '../data/final/'
```

```python
#Cargamos todos los datasets procesados
df_comercio = pd.read_csv(PROCESSED + 'df_comercio.csv')
df_poblacion = pd.read_csv(PROCESSED + 'df_poblacion.csv')
df_alquiler = pd.read_csv(PROCESSED + 'df_alquiler.csv')
df_catastral = pd.read_csv(PROCESSED + 'df_catastral.csv')

#Revisamos que están todos los registros correctamente
for nombre, df in [('Comercio', df_comercio), 
                    ('Población', df_poblacion),
                    ('Alquiler', df_alquiler), 
                    ('Catastral', df_catastral)]:
    print(f'{nombre}: {df.shape} | Nulos: {df.isnull().sum().sum()} | Barrios: {df.cod_barrio.nunique()}')
```

```python
# Verificar compatibilidad de códigos de barrio
cod_comercio = set(df_comercio['cod_barrio'].astype(str).str.zfill(3))
cod_poblacion = set(df_poblacion['cod_barrio'].astype(str).str.zfill(3))
cod_alquiler = set(df_alquiler['cod_barrio'].astype(str).str.zfill(3))
cod_catastral = set(df_catastral['cod_barrio'].astype(str).str.zfill(3))

print(f'Barrios en comercio:   {len(cod_comercio)}')
print(f'Barrios en población:  {len(cod_poblacion)}')
print(f'Barrios en alquiler:   {len(cod_alquiler)}')
print(f'Barrios en catastral:  {len(cod_catastral)}')
```

```python
#Unificamos todos los datasets
#Usamos cod_barrio como clave de unión

#Estandarizamos cod_barrio en todos los datasets
for df in [df_comercio, df_poblacion, df_alquiler, df_catastral]:
    df['cod_barrio'] = df['cod_barrio'].astype(str).str.zfill(3)

df_final = df_poblacion.merge(
    df_comercio.drop(columns=['barrio']), 
    on='cod_barrio', how='left')

df_final = df_final.merge(
    df_alquiler.drop(columns=['barrio']), 
    on='cod_barrio', how='left')

df_final = df_final.merge(
    df_catastral.drop(columns=['barrio']), 
    on='cod_barrio', how='left')

print(f'Dimensiones dataset final: {df_final.shape}')
print(f'Valores nulos:\n{df_final.isnull().sum()}')
print(f'\nColumnas disponibles:')
for i, col in enumerate(df_final.columns):
    print(f'  {i+1}. {col}')
```

```python
print('Estadísticas descriptivas')
print(df_final.describe().round(2).to_string())

print('\nMuestra 5 barrios representativos')
barrios_muestra = ['011', '012', '051', '121', '161']
print(df_final[df_final['cod_barrio'].isin(barrios_muestra)][
    ['barrio', 'pct_extranjera', 'pct_65mas','pct_hosteleria_2023', 'alquiler_2023', 'catastral_m2']].to_string())

df_final.to_csv(FINAL + 'dataset_barrios.csv', index=False)
```
