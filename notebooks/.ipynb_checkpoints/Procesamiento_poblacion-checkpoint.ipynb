```python
import pandas as pd
import numpy as np
import os

RAW = '../data/raw/'
PROCESSED = '../data/processed/'
FINAL = '../data/final/'
```

```python
#Cargamos los dos archivos de población
df_grandes = pd.read_excel(RAW + 'poblacion_grandes_grupos.xlsx', header=None)
df_quinquenal = pd.read_excel(RAW + 'poblacion_quinquenal.xlsx', header=None)

print('GRANDES GRUPOS')
print(f'Dimensiones: {df_grandes.shape}')
print('\nPrimeras 12 filas, primeras 5 columnas:')
print(df_grandes.iloc[:12, :5].to_string())

print('\nQUINQUENAL')
print(f'Dimensiones: {df_quinquenal.shape}')
print('\nPrimeras 12 filas, primeras 5 columnas:')
print(df_quinquenal.iloc[:12, :5].to_string())
```

```python
#Revisión cabeceras completas de ambos archivos
print('GRANDES GRUPOS- Cabeceras (filas 6-8)')
for i in range(6, 9):
    valores = [v for v in df_grandes.iloc[i, :].tolist() if v is not None and str(v) != 'nan']
    print(f'Fila {i}: {valores}')

print('\nQUINQUENAL- Cabeceras (filas 6-8)')
for i in range(6, 9):
    valores = [v for v in df_quinquenal.iloc[i, :].tolist() if v is not None and str(v) != 'nan']
    print(f'Fila {i}: {valores}')

#Vemos años disponibles en ambos
años_grandes = sorted(set(
    str(v).strip() for v in df_grandes.iloc[9:, 0] 
    if v is not None and 'enero' in str(v)
))
años_quinquenal = sorted(set(
    str(v).strip() for v in df_quinquenal.iloc[9:, 0] 
    if v is not None and 'enero' in str(v)
))

print(f'\nAños grandes grupos: {años_grandes}')
print(f'Años quinquenal: {años_quinquenal}')
```

```python
#Construimos nombres de columnas para grandes grupos
#Formato: nacionalidad_sexo_edad
cols_grandes = []
for i in range(12):
    nac = df_grandes.iloc[6, i+3]
    sexo = df_grandes.iloc[7, i+3]
    edad = df_grandes.iloc[8, i+3]
    cols_grandes.append(f'{nac}_{sexo}_{edad}')

print('Columnas grandes grupos:')
for i, c in enumerate(cols_grandes):
    print(f'  Col {i+3}: {c}')

#Construimos nombres de columnas para quinquenal
cols_quinquenal = []
for i in range(24):
    nac = df_quinquenal.iloc[6, i+3]
    sexo = df_quinquenal.iloc[7, i+3]
    edad = df_quinquenal.iloc[8, i+3]
    cols_quinquenal.append(f'{nac}_{sexo}_{edad}')

print('\nColumnas quinquenal:')
for i, c in enumerate(cols_quinquenal):
    print(f'  Col {i+3}: {c}')
```

```python
#Función para limpiar valores numéricos
def limpiar_numero(valor):
    if pd.isna(valor):
        return 0
    return int(str(valor).replace('.', '').replace(',', '').strip())

#Años que nos interesan
años_validos = ['1 de enero de 2018', '1 de enero de 2019', 
                '1 de enero de 2020', '1 de enero de 2021',
                '1 de enero de 2022', '1 de enero de 2023']

#Procesamos GRANDES GRUPOS
registros_grandes = []
for idx, row in df_grandes.iterrows():
    fecha = str(row[0]).strip()
    if fecha not in años_validos:
        continue
    
    barrio = str(row[2]).strip()
    #Solo los barrios reales (código de 3 dígitos)
    codigo = barrio.split('.')[0].strip()
    if not (len(codigo) == 3 and codigo.isdigit()):
        continue
    
    año = int(fecha.split(' ')[-1])
    registro = {'año': año, 'cod_barrio': codigo, 'barrio': barrio}
    
    for i, col_nombre in enumerate(cols_grandes):
        registro[col_nombre] = limpiar_numero(row[i+3])
    
    registros_grandes.append(registro)

df_g = pd.DataFrame(registros_grandes)
print(f'Grandes grupos - Dimensiones: {df_g.shape}')
print(f'Años: {sorted(df_g.año.unique())}')
print(f'Barrios: {df_g.cod_barrio.nunique()}')
print(df_g.head(3).to_string())
```

```python
#Procesamos QUINQUENAL
registros_quinquenal = []
for idx, row in df_quinquenal.iterrows():
    fecha = str(row[0]).strip()
    if fecha not in años_validos:
        continue
    
    barrio = str(row[2]).strip()
    codigo = barrio.split('.')[0].strip()
    if not (len(codigo) == 3 and codigo.isdigit()):
        continue
    
    año = int(fecha.split(' ')[-1])
    registro = {'año': año, 'cod_barrio': codigo, 'barrio': barrio}
    
    for i, col_nombre in enumerate(cols_quinquenal):
        registro[col_nombre] = limpiar_numero(row[i+3])
    
    registros_quinquenal.append(registro)

df_q = pd.DataFrame(registros_quinquenal)
print(f'Quinquenal - Dimensiones: {df_q.shape}')
print(f'Años: {sorted(df_q.año.unique())}')
print(f'Barrios: {df_q.cod_barrio.nunique()}')
print(df_q.head(3).to_string())
```

```python
#Unimos los dos dataframes por año, cod_barrio y barrio
df_pob = df_g.merge(df_q, on=['año', 'cod_barrio', 'barrio'])
print(f'Dataset unificado - Dimensiones: {df_pob.shape}')

#calculamos población total por barrio y año
df_pob['pob_total'] = (
    df_pob['Española_Hombres_0 a 15 años'] +
    df_pob['Española_Hombres_16 a 64 años'] +
    df_pob['Española_Hombres_65 y más años'] +
    df_pob['Española_Mujeres_0 a 15 años'] +
    df_pob['Española_Mujeres_16 a 64 años'] +
    df_pob['Española_Mujeres_65 y más años'] +
    df_pob['Extranjera_Hombres_0 a 15 años'] +
    df_pob['Extranjera_Hombres_16 a 64 años'] +
    df_pob['Extranjera_Hombres_65 y más años'] +
    df_pob['Extranjera_Mujeres_0 a 15 años'] +
    df_pob['Extranjera_Mujeres_16 a 64 años'] +
    df_pob['Extranjera_Mujeres_65 y más años'])

#calculamos población extranjera total
df_pob['pob_extranjera'] = (
    df_pob['Extranjera_Hombres_0 a 15 años'] +
    df_pob['Extranjera_Hombres_16 a 64 años'] +
    df_pob['Extranjera_Hombres_65 y más años'] +
    df_pob['Extranjera_Mujeres_0 a 15 años'] +
    df_pob['Extranjera_Mujeres_16 a 64 años'] +
    df_pob['Extranjera_Mujeres_65 y más años'])

#Calculamos población mayor de 65
df_pob['pob_65mas'] = (
    df_pob['Española_Hombres_65 y más años'] +
    df_pob['Española_Mujeres_65 y más años'] +
    df_pob['Extranjera_Hombres_65 y más años'] +
    df_pob['Extranjera_Mujeres_65 y más años'])

#Calculamos población 20-44 años (de los quinquenales)
df_pob['pob_2044'] = (
    df_pob['Española_Hombres_20 a 24 años'] +
    df_pob['Española_Hombres_25 a 29 años'] +
    df_pob['Española_Hombres_30 a 34 años'] +
    df_pob['Española_Hombres_35 a 39 años'] +
    df_pob['Española_Hombres_40 a 44 años'] +
    df_pob['Española_Mujeres_20 a 24 años'] +
    df_pob['Española_Mujeres_25 a 29 años'] +
    df_pob['Española_Mujeres_30 a 34 años'] +
    df_pob['Española_Mujeres_35 a 39 años'] +
    df_pob['Española_Mujeres_40 a 44 años'] +
    df_pob['Extranjera_Hombres_20 a 24 años'] +
    df_pob['Extranjera_Hombres_25 a 29 años'] +
    df_pob['Extranjera_Hombres_30 a 34 años'] +
    df_pob['Extranjera_Hombres_35 a 39 años'] +
    df_pob['Extranjera_Hombres_40 a 44 años'] +
    df_pob['Extranjera_Mujeres_20 a 24 años'] +
    df_pob['Extranjera_Mujeres_25 a 29 años'] +
    df_pob['Extranjera_Mujeres_30 a 34 años'] +
    df_pob['Extranjera_Mujeres_35 a 39 años'] +
    df_pob['Extranjera_Mujeres_40 a 44 años'])

#Calculamos población extranjera 20-44
df_pob['pob_extranjera_2044'] = (
    df_pob['Extranjera_Hombres_20 a 24 años'] +
    df_pob['Extranjera_Hombres_25 a 29 años'] +
    df_pob['Extranjera_Hombres_30 a 34 años'] +
    df_pob['Extranjera_Hombres_35 a 39 años'] +
    df_pob['Extranjera_Hombres_40 a 44 años'] +
    df_pob['Extranjera_Mujeres_20 a 24 años'] +
    df_pob['Extranjera_Mujeres_25 a 29 años'] +
    df_pob['Extranjera_Mujeres_30 a 34 años'] +
    df_pob['Extranjera_Mujeres_35 a 39 años'] +
    df_pob['Extranjera_Mujeres_40 a 44 años'])

cols_check = ['año', 'cod_barrio', 'barrio', 'pob_total','pob_extranjera', 'pob_65mas', 'pob_2044', 'pob_extranjera_2044']
print(df_pob[cols_check].head(5).to_string())
```

```python
#Calculamos indicadores como porcentaje sobre población total
df_pob['pct_extranjera'] = (df_pob['pob_extranjera'] / df_pob['pob_total'] * 100).round(2)

df_pob['pct_65mas'] = (df_pob['pob_65mas'] / df_pob['pob_total'] * 100).round(2)

df_pob['pct_pob_2044'] = (df_pob['pob_2044'] / df_pob['pob_total'] * 100).round(2)

df_pob['pct_extranjera_2044'] = (df_pob['pob_extranjera_2044'] / df_pob['pob_total'] * 100).round(2)

#Separamos 2018 y 2023 para calcular variaciones
df_pob_2018 = df_pob[df_pob['año'] == 2018][
    ['cod_barrio', 'pct_extranjera', 'pct_65mas','pct_pob_2044', 'pct_extranjera_2044']
].rename(columns={
    'pct_extranjera': 'pct_extranjera_2018',
    'pct_65mas': 'pct_65mas_2018',
    'pct_pob_2044': 'pct_pob_2044_2018',
    'pct_extranjera_2044': 'pct_extranjera_2044_2018'})

df_pob_2023 = df_pob[df_pob['año'] == 2023][
    ['cod_barrio', 'barrio', 'pob_total',
     'pct_extranjera', 'pct_65mas', 
     'pct_pob_2044', 'pct_extranjera_2044']]

#Unimos y calculamos variaciones
df_pob_final = df_pob_2023.merge(df_pob_2018, on='cod_barrio')

df_pob_final['var_extranjera_1823'] = (df_pob_final['pct_extranjera'] - df_pob_final['pct_extranjera_2018']).round(2)

df_pob_final['var_65mas_1823'] = (df_pob_final['pct_65mas'] - df_pob_final['pct_65mas_2018']).round(2)

df_pob_final['var_pob_2044_1823'] = (df_pob_final['pct_pob_2044'] - df_pob_final['pct_pob_2044_2018']).round(2)

df_pob_final['var_extranjera_2044_1823'] = (df_pob_final['pct_extranjera_2044'] - df_pob_final['pct_extranjera_2044_2018']).round(2)

print(f'Dimensiones dataset población: {df_pob_final.shape}')
print(f'Valores nulos:\n{df_pob_final.isnull().sum()}')

cols_mostrar = ['cod_barrio', 'barrio', 'pct_extranjera', 'pct_65mas','pct_pob_2044', 'pct_extranjera_2044','var_extranjera_1823', 'var_pob_2044_1823']
print(df_pob_final[cols_mostrar].head(10).to_string())
```

```python
#Seleccionar columnas finales para el dataset de población
df_poblacion_final = df_pob_final[[
    'cod_barrio',
    'barrio',
    'pob_total',
    'pct_extranjera',
    'pct_65mas',
    'pct_pob_2044',
    'pct_extranjera_2044',
    'var_extranjera_1823',
    'var_65mas_1823',
    'var_pob_2044_1823',
    'var_extranjera_2044_1823'
]].copy()

print(f'Dimensiones finales: {df_poblacion_final.shape}')
print(f'\nEstadísticas 2023:')
print(df_poblacion_final.describe().round(2))

df_poblacion_final.to_csv(PROCESSED + 'df_poblacion.csv', index=False)
```
